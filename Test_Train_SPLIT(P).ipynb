```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/drive/MyDrive/P3 Fitrec Dataset/Generated datsets fresh/ClusteredData_01-11-2023.csv')
```

```python
# Assuming you have a DataFrame named 'df'

# Step 1: Shuffle the rows
shuffled_df = df.sample(frac=1, random_state=42)  # Shuffle with a fixed random seed for reproducibility
```

```python
shuffled_df.head()
```

```python
# Step 2: Split the shuffled DataFrame into training and test DataFrames (80/20 split)
train_ratio = 0.80
test_ratio = 0.20

train_df, test_df = train_test_split(shuffled_df, test_size=test_ratio, random_state=42)

# 'train_df' contains 80% of the data, and 'test_df' contains 20% of the data

# Optionally, reset the index of the DataFrames if needed
train_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)
```
